```python
import numpy as np

# 定义参数
N = 100  # 采样点数
Ts = 0.01  # 采样周期
n = np.arange(N)
x = np.sin(2 * np.pi * n * Ts)  # 采样信号


# DFT
X = np.fft.rfft(x)


# 频域截断
k1 = 0  # 起始频率索引
k2 = 30  # 结束频率索引
H = np.zeros_like(X, dtype=int)
H[k1:k2] = 1
Y = X * H


# IDFT
y = np.fft.irfft(Y)


# print("Original signal:", x)
# print("Truncated and reconstructed signal:", np.real(y))
```

```python
import matplotlib.pyplot as plt

plt.plot(X)
# plt.plot(y)
plt.show()
```

```python
# 计算rmse
mse = np.sqrt(np.mean((x - np.real(y)) ** 2))
print("MSE:", mse)
```
